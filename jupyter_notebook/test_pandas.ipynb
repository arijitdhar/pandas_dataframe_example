Indexing 

```python
df6
```

```python
# Label based indexing
# Pass Index range and column range
# df6.loc["1":"2","Country":"Employee"]
# You can pass single index as well
df6.loc[1,"Country"]
df6.loc["1":"2","Country":"Employee"]
```

```python
# Position based indexing
# DataFrame.iloc-> First argument is position of Index, Second Argument is position of Column
df6.iloc[0,4]
```

```python
# Remember, in list range, upper bound is exclusive
df6.iloc[0:4, 0:4]
```

Deleting columns and rows

```python
# Use Drop method, creates a new DF, so not an inplace operation
# 1-> To Drop Column
# 0-> To Drop Row
df6.drop("City", 1)
```

```python
# To Drop Row, pass the index value of that row and the second argument as 0, to signify it as a row drop operation
df6.drop(2, 0)
```

```python
# Delete row based on position(index), trick is to get the value of the index, using df.index[position]
df6.drop(df6.index[1], 0)
```

```python
# Delete column based on position(index), trick is to get the value of the column, using df.columns[position]
df6.drop(df6.columns[2], 1)
```

Updating and adding new columns and rows

```python
len(df6.index)
# Shape returns the matrix size, in this case (6,7)
# df6.shape
# Add a new column, use shape to get the number of rows and multiply with that to have the same value for the 
# new column for all the rows
# The below is an in-place operation
df6["Continent"] = df6.shape[0]*["North America"]
df6
```

```python
# Update a column
df6["Continent"] = df6["Country"]+ "," + "North America"
df6
```

```python
# Update a row
# There is no direct way to add a new row, we have to transpose the matrix and add a column and then transpose the matrix again
df6_t = df6.T
df6_t
```

```python
# In order to update an existing row, pass an existing value, for addition, pass a new value as below: df6_t["7"]
df6_t["7"] = ["7","My Address", "My City", "700047", "India", "My Shop", 10, "Asia"]
df6_t
```

```python
df6 = df6_t.T
df6
```
